```python
import geopandas as gpd
units=gpd.read_file("spatial_units.gpkg", driver="GPKG")
```

```python
unit=units.iloc[0]
print(unit)
unit_gdf = gpd.GeoDataFrame(
[unit], crs=units.crs)
unit_gdf.geometry = unit_gdf.geometry.centroid
unit_gdf.to_crs(epsg=4326, inplace=True)
```
